# Lecture 1 - Lifting line theory

This lecture is about lifting line theory.
From thin airfoil theory, we have that the lift of a symmetric airfoil is given by the following formula.

```python
from math import pi
def thin_lift(α):
    return 2*pi*α
```

The effective angle of attack $\alpha_{\mathrm{e}}$ is given by the formula

$$
\alpha_{\mathrm{e}} = \alpha - \alpha_{\mathrm{i}},
$$

where $\alpha_{\mathrm{i}}$ is the induced angle og attack.

The dynamic pressure $p_{\mathrm{d}}$ is given by the formula

$$
p_{\mathrm{d}} = \frac{\varrho U^2}{2},
$$

from which we can find the lift through

$$
L = p_{\mathrm{d}} C_L S
$$

Inverting the formula

$$
L = \Gamma \varrho U,
$$

we can find the circulation $\Gamma$:

```python
from math import radians
b = 1
AR = 5
C = b/AR
ϱ = 1.25
U = 10
p_dynamic = .5*ϱ*U**2
def circulation(α, α_induced):
    α_effective = α - α_induced
    C_L = thin_lift(α_effective)
    L = p_dynamic * C_L * C
    return L/(ϱ*U)
print(f"Γ: {circulation(radians(5), 0)}")
```
